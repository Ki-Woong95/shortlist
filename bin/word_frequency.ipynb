```python
C_cond = []
CV_cond = []
CVC_cond = []
for i in sequence:
    C1, V1, C2 = i.split(' ')
    if C1 not in C_cond:
        C_cond.append(C1)
    if C1 + ' ' + V1 not in CV_cond:
        CV_cond.append(C1 + ' ' + V1)
    if C1 + ' ' + V1 + ' ' + C2 not in CVC_cond:
        CVC_cond.append(C1 + ' ' + V1 + ' ' + C2)
```

```python
import csv
import nltk
from nltk.corpus import cmudict

# Download the CMU Pronouncing Dictionary
nltk.download('cmudict')

# Load the CMU Pronouncing Dictionary
pronouncing_dict = cmudict.dict()

def remove_stress(phoneme):
    return ''.join(char for char in phoneme if char.isalpha())
def count_words_matching_condition(condition, cmudict):
    count = 0
    for word, pronunciations in cmudict.items():
        for pronunciation in pronunciations:
            phoneme_sequence = ' '.join(remove_stress(phoneme) for phoneme in pronunciation)
            if condition in phoneme_sequence:
                count += 1
    return count
```

```python
# Write results to CSV file
many = ['P', 'K', 'S']
few = ['JH', 'CH', 'TH', 'V', 'SH']
csv_filename = '../data/amer _word_counts_v2.csv'
with open(csv_filename, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)

    # Write header
    csv_writer.writerow(['Condition', 'Word Count', 'Structure', 'Competitors'])

    # Write results for C_cond
    for condition in C_cond:
        word_count = count_words_matching_condition(condition, pronouncing_dict)
        c1  = condition.split(' ')[0]
        if c1 in many:
            csv_writer.writerow([condition, word_count, 'C', 'Many'])
        else:
            csv_writer.writerow([condition, word_count, 'C', 'Few'])

    # Write results for CV_cond
    for condition in CV_cond:
        word_count = count_words_matching_condition(condition, pronouncing_dict)
        c1  = condition.split(' ')[0]
        if c1 in many:
            csv_writer.writerow([condition, word_count, 'CV', 'Many'])
        else:
            csv_writer.writerow([condition, word_count, 'CV', 'Few'])

    # Write results for CVC_cond
    for condition in CVC_cond:
        word_count = count_words_matching_condition(condition, pronouncing_dict)
        c1  = condition.split(' ')[0]
        if c1 in many:
            csv_writer.writerow([condition, word_count, 'CVC', 'Many'])
        else:
            csv_writer.writerow([condition, word_count, 'CVC', 'Few'])


print(f"Results written to {csv_filename}")
```

## Separate C, CV, CVC condition per each experiment

Main objective: replace many/few condition into continuous variables (frequency of each condition) except for C only condition:
C: Many / Few (categorical)
CV & CVC: Frequency of each sequence as predictor for competitors

## American

```python
#Load frequency data
american = pd.read_csv('../data/amer _word_counts_v2.csv')
american.head()
```

```python
#Load results of the experiment
am_result = pd.read_csv('../data/american.csv')
am_res_correct = am_result[am_result['isCorrect'] == 1]
am_res_correct['targetWord'][:5]
am_res_correct.head()
```

```python
#Load stimuli & phoneme sequence for mapping
am_sti = pd.read_csv('../data/amer_stimuli.csv')
am_ipa = list(am_sti['Stimuli'])
am_arpa = list(am_sti['Phoneme Sequence'])
print(am_ipa)
```

```python
res_sti = list(set(am_res_correct['Stimuli[Trial]']))
ipa_arpa = {}
for i in range(len(am_sti)):
    ipa_arpa[am_sti['Stimuli'][i]] = am_sti['Phoneme Sequence'][i]
len(ipa_arpa)
```

```python
# Define a function to map IPA to ARPABET
def map_ipa_to_arpabet(ipa):
    return ipa_arpa.get(ipa, None)

# Apply the mapping function to create the 'sequence' column using .loc
am_res_correct.loc[:, 'sequence'] = am_res_correct['Stimuli[Trial]'].map(map_ipa_to_arpabet)
```

```python
am_res_correct.head()
```

### 1. CVC condition (American)

```python
freq_dic = {}
for i in range(len(american)):
    freq_dic[american['Condition'][i]] = american['Word Count'][i]
```

```python
CVC_am = am_res_correct.copy()
def CVC_count(seq):
    return freq_dic.get(seq, None)

CVC_am.loc[:, 'CVC_freq'] = CVC_am['sequence'].map(CVC_count)
```

```python
CVC_am
```

```python
CVC_am.to_csv('../data/freq_count_by_structure/American/CVC_condition.csv', index = False)
```

### 2. CV Condition (American)

```python
def remove_last_consonant(sequence):
    # Split the sequence by whitespace
    sequence_list = sequence.split()
    # Remove the last element (consonant)
    sequence_list = sequence_list[:-1]
    # Join the remaining elements back together
    return ' '.join(sequence_list)
```

```python
CV_am = am_res_correct.copy()
CV_am['sequence'] = CV_am['sequence'].apply(remove_last_consonant)
```

```python
def CV_count(seq):
    return freq_dic.get(seq, None)

CV_am.loc[:, 'CV_freq'] = CV_am['sequence'].map(CV_count)
```
